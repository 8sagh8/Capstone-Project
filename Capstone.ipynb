# MODEL

```python
data.head(1)
```

```python
#Cleaning text in description series

import regex as re

df['description'] = df['description'].str.lower()
df['description'] = df['description'].apply(lambda x: re.sub(r'[^\w\d\s\']+', '', x))

df.head()
```

```python
#Tokenizing description series
import nltk 
from nltk.tokenize import word_tokenize

df['description_token'] = df['description'].apply(word_tokenize)
df.head(1)
```

```python
#remove stopwords from description
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
df['description_clean'] =  df['description_token'].apply(lambda x: [word for word in x if word not in stop_words])
df.head(1)
```

```python
#join string back for analysis

def combine_words(words_list):
    # Combine the words into a single string with spaces
    combined_string = " ".join(words_list)
    
    return combined_string


df["Total"] = df["description_clean"].apply(combine_words)

# Print the DataFrame
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer

transformer = TfidfTransformer(smooth_idf = False)
count_vectorizer = CountVectorizer(ngram_range = (1,2))
```

```python
train_counts = count_vectorizer.fit_transform(df['Total'].values)

train_tfidf = transformer.fit_transform(train_counts)
train_counts
```

```python
X = train_tfidf
y = df['fraudulent']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 3)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

Adab = AdaBoostClassifier(DecisionTreeClassifier(max_depth = 3), n_estimators = 5)
Adab.fit(X_train, y_train)
```

```python
Adab.score(X_train, y_train)
```

```python
Adab.score(X_test, y_test)
```

## DATA IMBALANCE

```python
print('The number of Non-Frauds are: ' + str(data['fraudulent'].value_counts()[0]) + ' which is ', round(data['fraudulent'].value_counts()[0]/len(data) * 100,2), '% of the dataset')
print('The number of Frauds are: ' + str(data['fraudulent'].value_counts()[1]) + ' which is ', round(data['fraudulent'].value_counts()[1]/len(data) * 100,2), '% of the dataset')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Shuffle the Dataset.
shuffled_df = data.sample(frac=1,random_state=4)

# Put all the fraud class in a separate dataset.
fraud_df = data.loc[shuffled_df['fraudulent'] == 1]

#Randomly select 492 observations from the non-fraud (majority class)
non_fraud_df = shuffled_df.loc[shuffled_df['fraudulent'] == 0].sample(n=492,random_state=42)

# Concatenate both dataframes again
normalized_df = pd.concat([fraud_df, non_fraud_df])

#plot the dataset after the undersampling
plt.figure(figsize=(8, 8))
sns.countplot('fraudulent', data=normalized_df)
plt.title('Balanced Classes')
plt.show()
```
