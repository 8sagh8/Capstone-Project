````python
import regex as re
import regex as re

df['title'] = df['title'].str.lower()
df['title'] = df['title'].apply(lambda x: re.sub(r'[^\w\d\s\']+', '', x))

df['description'] = df['description'].str.lower()
df['description'] = df['description'].apply(lambda x: re.sub(r'[^\w\d\s\']+', '', x))

df['requirements'] = df['requirements'].str.lower()
df['requirements'] = df['requirements'].apply(lambda x: re.sub(r'[^\w\d\s\']+', '', x))

df['benefits'] = df['benefits'].str.lower()
df['benefits'] = df['benefits'].apply(lambda x: re.sub(r'[^\w\d\s\']+', '', x))

df['employment_type'] = df['employment_type'].str.lower()
df['employment_type'] = df['employment_type'].apply(lambda x: re.sub(r'[^\w\d\s\']+', '', x))

df['required_experience'] = df['required_experience'].str.lower()
df['required_experience'] = df['required_experience'].apply(lambda x: re.sub(r'[^\w\d\s\']+', '', x))

df['required_education'] = df['required_education'].str.lower()
df['required_education'] = df['required_education'].apply(lambda x: re.sub(r'[^\w\d\s\']+', '', x))

df['industry'] = df['industry'].str.lower()
df['industry'] = df['industry'].apply(lambda x: re.sub(r'[^\w\d\s\']+', '', x))
````

````python

#Tokenizing description series
import nltk 
from nltk.tokenize import word_tokenize
nltk.download('punkt')

df['title_token'] = df['title'].apply(word_tokenize)

df['description_token'] = df['description'].apply(word_tokenize)

df['requirements_token'] = df['requirements'].apply(word_tokenize)

df['benefits_token'] = df['benefits'].apply(word_tokenize)

df['employment_type_token'] = df['employment_type'].apply(word_tokenize)

df['required_experience_token'] = df['required_experience'].apply(word_tokenize)

df['required_education_token'] = df['required_education'].apply(word_tokenize)

df['industry_token'] = df['industry'].str.lower()

df.head(2)
````

````python

#remove stopwords from description
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

df['title_clean'] = df['title_token'].apply(lambda x: [word for word in x if word not in stop_words])

df['description_clean'] = df['description_token'].apply(lambda x: [word for word in x if word not in stop_words])

df['requirements_clean'] = df['requirements_token'].apply(lambda x: [word for word in x if word not in stop_words])

df['benefits_clean'] = df['benefits_token'].apply(lambda x: [word for word in x if word not in stop_words])

df['employment_type_clean'] = df['employment_type_token'].apply(lambda x: [word for word in x if word not in stop_words])

df['required_experience_clean'] = df['required_experience_token'].apply(lambda x: [word for word in x if word not in stop_words])

df['required_education_clean'] = df['required_education_token'].apply(lambda x: [word for word in x if word not in stop_words])

df['industry_clean'] = df['industry_token'].apply(lambda x: [word for word in x if word not in stop_words])

df.head(2)
````

````python

#join string back for analysis

def combine_words(words_list):
    # Combine the words into a single string with spaces
    combined_string = " ".join(words_list)
    
    return combined_string


df["Total"] = df["description_clean"].apply(combine_words)

df['title_Total'] = df['title_clean'].apply(combine_words)

# df['description_Total'] = df['description_clean'].apply(combine_words)

df['requirements_Total'] = df['requirements_clean'].apply(combine_words)

df['benefits_Total'] = df['benefits_clean'].apply(combine_words)

df['employment_type_Total'] = df['employment_type_clean'].apply(combine_words)

df['required_experience_Total'] = df['required_experience_clean'].apply(combine_words)

df['required_education_Total'] = df['required_education_clean'].apply(combine_words)

df['industry_Total'] = df['industry_clean'].apply(combine_words)


# Print the DataFrame
df.head()
````
