# Fill missing values to Function 

```python
# function will match value of function/department with title
# if value found it will be returned that value OR returned False
def match_word(mylist, mytext):
    for word in mylist:
        if word.lower() in mytext.lower():
            return word.lower()
    return False

# To return all the values in the column Function / Job_Category
def getFunctionIndex(df):
    fi = df.function.value_counts().index
    return fi
```

```python
ddd = pd.read_csv('./newFile.csv')
# ddd.info()
```

```python
size = ddd.shape[0]


for i in range(0, size):
    if pd.isnull(ddd.function[i]):
        fi = getFunctionIndex(ddd)
        match = match_word(fi, ddd.title[i])
        if (match == False):
            print ("Enter New Function, Job Category")
            print("index is:", i)
            print("Title is:", ddd.title[i])
            c = input("Enter here!")
            ddd.function[i] = c
        else:
            ddd.function[i] = match
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("./newFile.csv")
df.head()
```

```python
data = df.copy()
```

```python
data.head()
```

```python
import re

data['description']= data['description'].str.replace('\n',' ')
data['description']= data['description'].str.replace('\r',' ')
data['description']= data['description'].str.replace('\t',' ')
data['description'] = data['description'].apply(lambda x: re.sub(r'[0-9]',' ',x))
data['description'] = data['description'].apply(lambda x: re.sub(r'[/(){}\[\]\|@,;.:-]',' ',x))
data['description']= data['description'].apply(lambda s:s.lower() if type(s) == str else s)
data['description']= data['description'].str.replace('  ',' ')

import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords

print(stopwords.words("english"))

stop_words = set(stopwords.words("english"))
data['text'] = data['text'].apply(lambda x:' '.join([word for word in x.split() if word not in (stop_words)]))
data['text'][0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer(max_features=2000)
df = pd.DataFrame(tf.fit_transform(data['description']).toarray(),columns=tf.get_feature_names_out())
data.drop(['description'],axis=1,inplace=True)
x_df = pd.concat([df,data],axis=1)
```
